## Image recognition of Currency using Support Vector Machines


Hi guys!
In this python notebook we are going to see about image recognition of country currency using Support Vector Machine or SVM.
Support Vector Machines or SVM is one of the Supervised Learning Algorithm. SVM is used in both Regression and  as well as Classification.
SVM draws a decision boundary which is a hyperplane between any two classes in order to separate them or classify them.
Support vectors are the data points, which are closest to the hyperplane. These points will define the separating line better by calculating margins. These points are more relevant to the construction of the classifier.

### Assigning training set and test set

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

### Perform Principle Component Analysis

Principle Component Analysis or PCA is a Feature Extraction Techniques.
Principal Component Analysis is an unsupervised learning algorithm that is used for the dimensionality reduction in machine learning.

```python
from sklearn import decomposition
```

```python
pca=decomposition.PCA(n_components=90,whiten=True,random_state=1)
```

```python
pca.fit(x_train)
```

```python
x_train_pca=pca.transform(x_train)
```

```python
x_test_pca=pca.transform(x_test)
```

```python
eigen=(np.reshape(x[100],(600,600)).astype(np.float))
```

```python
fig=plt.figure(figsize=(30,30))
for i in range(90):
    ax=fig.add_subplot(9,10,i+1,xticks=[],yticks=[])
    ax.imshow(pca.components_[i].reshape(eigen.shape),
             cmap=plt.cm.Spectral)
    
```

```python
from sklearn import svm              #svm-support vector machine
```

```python
img_clf=svm.SVC(C=2.0,gamma=0.006,kernel='rbf',random_state=42)
```

```python
img_clf.fit(x_train_pca,y_train)
```

```python
y_pred=img_clf.predict(x_test_pca)
```
